```python
import numpy as np
import matplotlib.pyplot as plt
# Plotting area

plt.axis([0,100,0,10])
plt.axis('on')
plt.grid(True)
plt.title('ini adalah plot saya')
plt.xlabel('ini adalah x axis')
plt.ylabel('ini adalah y axis')

# Color intensity
plt.axis([0,100,0,10])
plt.scatter(60,50,s=1000,color='b',alpha=1)
plt.scatter(80,50,s=1000,color='b',alpha=.5)
plt.scatter(100,50,s=1000,color='b',alpha=.1)
plt.show()
```
